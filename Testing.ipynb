```python
import google.generativeai as genai
import PIL 
from PIL import Image
import os
```

```python
defect_image = Image.open('structure damage image.jpg')
```

```python
defect_image
```

```python
gemini_api_key = os.getenv('Google_api_key1')
genai.configure(api_key = gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = 'Identify the structural defect in the image and suggest how it can be fixed.'
```

```python
response = model.generate_content([prompt,defect_image],generation_config={'temperature':0.7})
```

```python
print(response.text)
```
